```python

import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, Model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
import os
import cv2
from tqdm import tqdm

# ==================== 1. CREATE DUMMY DATASET ====================
print("Step 1: Creating sample dataset...")

# Create synthetic housing data
np.random.seed(42)
n_samples = 1000

# Tabular features: area, bedrooms, bathrooms, location_score
tabular_data = pd.DataFrame({
    'area': np.random.randint(500, 3500, n_samples),
    'bedrooms': np.random.randint(1, 5, n_samples),
    'bathrooms': np.random.randint(1, 4, n_samples),
    'location_score': np.random.uniform(1, 10, n_samples),
    'price': np.random.randint(100000, 500000, n_samples)  # target
})

# ==================== 2. CREATE SYNTHETIC IMAGES ====================
print("Step 2: Creating synthetic images...")

# Create directory for images
os.makedirs('house_images', exist_ok=True)

# Generate random images (64x64 RGB) for each house
image_data = []
for i in tqdm(range(n_samples)):
    # Create a synthetic house image (just random patterns for demo)
    img = np.random.randint(0, 255, (64, 64, 3), dtype=np.uint8)
    # Add some pattern to make it look like a house (simple shapes)
    cv2.rectangle(img, (20, 30), (44, 50), (100, 100, 100), -1)  # "house"
    cv2.rectangle(img, (30, 38), (34, 42), (200, 200, 200), -1)  # "door"

    # Save image
    img_path = f'house_images/house_{i}.jpg'
    cv2.imwrite(img_path, img)
    image_data.append(img_path)

tabular_data['image_path'] = image_data

# ==================== 3. LOAD AND PREPROCESS DATA ====================
print("\nStep 3: Loading and preprocessing...")

# Load images
def load_images(image_paths, target_size=(64, 64)):
    images = []
    for path in tqdm(image_paths):
        img = cv2.imread(path)
        img = cv2.resize(img, target_size)
        img = img / 255.0  # normalize
        images.append(img)
    return np.array(images)

# Split data
X_tab = tabular_data[['area', 'bedrooms', 'bathrooms', 'location_score']].values
y = tabular_data['price'].values
image_paths = tabular_data['image_path'].values

# Train-test split
X_tab_train, X_tab_test, y_train, y_test, img_paths_train, img_paths_test = train_test_split(
    X_tab, y, image_paths, test_size=0.2, random_state=42
)

# Scale tabular data
scaler = StandardScaler()
X_tab_train_scaled = scaler.fit_transform(X_tab_train)
X_tab_test_scaled = scaler.transform(X_tab_test)

# Load images
X_img_train = load_images(img_paths_train)
X_img_test = load_images(img_paths_test)

print(f"Tabular train shape: {X_tab_train_scaled.shape}")
print(f"Image train shape: {X_img_train.shape}")

# ==================== 4. BUILD MULTIMODAL MODEL ====================
print("\nStep 4: Building multimodal model...")

# Image input branch (CNN)
image_input = layers.Input(shape=(64, 64, 3), name='image_input')
x = layers.Conv2D(32, (3, 3), activation='relu')(image_input)
x = layers.MaxPooling2D((2, 2))(x)
x = layers.Conv2D(64, (3, 3), activation='relu')(x)
x = layers.MaxPooling2D((2, 2))(x)
x = layers.Conv2D(64, (3, 3), activation='relu')(x)
x = layers.Flatten()(x)
x = layers.Dense(64, activation='relu')(x)
image_output = layers.Dropout(0.5)(x)

# Tabular input branch
tabular_input = layers.Input(shape=(4,), name='tabular_input')
y = layers.Dense(32, activation='relu')(tabular_input)
y = layers.Dense(16, activation='relu')(y)
tabular_output = y

# Combine both branches
combined = layers.concatenate([image_output, tabular_output])
z = layers.Dense(64, activation='relu')(combined)
z = layers.Dense(32, activation='relu')(z)
z = layers.Dropout(0.3)(z)
output = layers.Dense(1, name='price_output')(z)

# Create model
model = Model(inputs=[image_input, tabular_input], outputs=output)

# Compile model
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

print(model.summary())

# ==================== 5. TRAIN MODEL ====================
print("\nStep 5: Training model...")

history = model.fit(
    [X_img_train, X_tab_train_scaled],
    y_train,
    validation_split=0.2,
    epochs=10,  # Increase for better results
    batch_size=32,
    verbose=1
)

# ==================== 6. EVALUATE ====================
print("\nStep 6: Evaluating model...")

# Predict
y_pred = model.predict([X_img_test, X_tab_test_scaled])

# Calculate metrics
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"\n📊 Results:")
print(f"Mean Absolute Error (MAE): ${mae:,.2f}")
print(f"Root Mean Square Error (RMSE): ${rmse:,.2f}")

# Plot training history
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Model Loss')

plt.subplot(1, 2, 2)
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual vs Predicted')

plt.tight_layout()
plt.savefig('multimodal_results.png')
plt.show()

print("\n✅ Task 3 Complete!")
```
